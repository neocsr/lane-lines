```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
m1 = 0.479621
b1 = 107

m2 = -0.638936
b2 = 845

x = np.linspace(0, 960)
```

```python
x.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
plt.plot(x, m1 * x + b1)
plt.plot(x, m2 * x + b2)
plt.axis([0, 960, 0, 540])
plt.gca().invert_yaxis()
ax.xaxis.set_label_position('top')
plt.show()
```

```python
m1 = 0.479621
b1 = 107
m2 = -0.638936
b2 = 845
x = np.linspace(0, 960)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
plt.plot(x, m1 * x + b1)
plt.plot(x, m2 * x + b2)
plt.plot(x, (0.52/0.85) * x + b1)
plt.axis([0, 960, 0, 540])
plt.gca().invert_yaxis()
ax.xaxis.set_label_position('top')
plt.show()
```
